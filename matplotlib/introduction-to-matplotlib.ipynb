## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # (width, height of the plot)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("./sample-plot.png")
```

## Making figures with NumPy arrays

We want:
* Line Plot
* Scatter Plot
* Bar Plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100) #between 0, 10, with 100 as size.
x
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use the same data to make scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary

#prepare the data
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
nut_butter_prices
```

```python
fig, ax = plt.subplots()
x = nut_butter_prices.keys()
y = nut_butter_prices.values()
store_name = "Dan's Butter"
y_axis_label = "Price ($)"
ax.bar(x,y);

# I can use f string here
ax.set(title = f"{store_name} Store",
       ylabel=y_axis_label); 
```

```python
#https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.barh.html
#check the documentation for the bar graph

fig, ax = plt.subplots()
ax.set(title="Dan's Butter Store", xlabel="Prices ($)")
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()), color='orange', align='center');
```

```python
#Make some data for histograms and plot it
x = np.random.randn(1000)

fig, ax = plt.subplots()
ax.hist(x);
```
